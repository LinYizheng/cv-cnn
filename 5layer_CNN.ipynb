```python
#
def conv_bn_pool( x, conv_size, channels, out_channels,cstride_size, wb_initializer, activation,
                 pool_type= None,pool_size= None,pstride_size= None, is_training= None):
    # Create variable named "weights".
    kernel_shape, bias_shape = [conv_size,conv_size,channels,out_channels],[out_channels]
    weights = tf.get_variable(
        name='conv_weights', shape=kernel_shape,initializer=wb_initializer)
    # Create variable named "biases".
    biases = tf.get_variable(
        name='conv_bias', shape=bias_shape,initializer=wb_initializer)
    conv = tf.nn.conv2d(
        x, weights, strides=[1, cstride_size, cstride_size, 1], padding='SAME')
    z = conv + biases
    
    if is_training != None:
        bn = batch_normalization(z,is_training)
        cnn_out = activation(bn)
    else:
        cnn_out = activation(z)
        
    if pool_type != None:
        cnn_out = pool_type(
            cnn_out, ksize=[1, pool_size, pool_size, 1],strides=[1, pstride_size, pstride_size, 1], padding='SAME')

    return cnn_out


def batch_normalization(x,is_training):
    x_shape = x.get_shape()
    params_shape = x_shape[-1]
    axis = list(range(len(x_shape) - 1))
    #The bias for batch normalization
    beta = tf.get_variable(
        name='beta',shape=params_shape,initializer=tf.zeros_initializer)
    #The scale of batch normalization
    gamma = tf.get_variable(
        name='gamma',shape=params_shape,initializer=tf.ones_initializer)
    #Record moving average for testing
    moving_mean = tf.get_variable(
        name='moving_mean',shape=params_shape,initializer=tf.zeros_initializer,trainable=False)
    #Record moving average for testing
    moving_variance = tf.get_variable(
        name='moving_variance',shape=params_shape,initializer=tf.ones_initializer,trainable=False)  
    # These ops will only be preformed when training.
    mean, variance = tf.nn.moments(x, axis)
    #Update moving averages
    decay = 0.95
    update_moving_mean = tf.assign(moving_mean,moving_mean * decay + mean * (1 - decay))
    update_moving_variance = tf.assign(moving_variance,moving_variance * decay + variance * (1 - decay))
    if is_training=='True':
        x = tf.nn.batch_normalization(x, mean, variance, beta, gamma, 0.001)
    else:
        with tf.control_dependencies([update_moving_mean, update_moving_variance]):
            x = tf.nn.batch_normalization(x, moving_mean, moving_variance, beta, gamma, 0.001) 
    return x

initializers = [tf.constant_initializer(0),tf.truncated_normal_initializer(0,0.1),tf.random_normal_initializer(0,0.1),
                                       tf.random_uniform_initializer(-0.1,0.1),tf.uniform_unit_scaling_initializer(),
                                       tf.glorot_uniform_initializer(),tf.glorot_normal_initializer()]
activations = [tf.nn.sigmoid,tf.nn.tanh,tf.nn.relu,tf.nn.softplus,tf.nn.elu,tf.nn.selu,tf.nn.relu6]
optimizers = ['Adam','GradientDescend','Adagrad','Momentum','Ftrl','RMSProp']

space = {
    'conv_size1': hp.choice('conv_size1',range(3,11)),
    'out_channels1': hp.choice('out_channels1',range(10,201)),                
    'cstride_size1': hp.choice('cstride_size1',range(2,4)),
    'wb_initializer1': hp.choice('wb_initializer1',initializers),
    'activation1': hp.choice('activation1', activations),
    'pool_type1': hp.choice('pool_type1',[tf.nn.max_pool,tf.nn.avg_pool]),                    
    'pool_size1': hp.choice('pool_size1', range(2,4)),                      
    'pstride_size1': hp.choice('pstride_size1',range(1,3)), 

    'conv_size2': hp.choice('conv_size2', range(3,8)),
    'out_channels2': hp.choice('out_channels2',range(10,201)),                
    'cstride_size2': hp.choice('cstride_size2',range(1,3)),
    'wb_initializer2': hp.choice('wb_initializer2',initializers),
    'activation2': hp.choice('activation2', activations),
    'pool_type2': hp.choice('pool_type2',[tf.nn.max_pool,tf.nn.avg_pool]),                    
    'pool_size2': hp.choice('pool_size2',range(2,4)),                      
    'pstride_size2': hp.choice('pstride_size2',range(1,3)),  
    
    'conv_size3': hp.choice('conv_size3', range(3,6)),
    'out_channels3': hp.choice('out_channels3',range(10,201)),                
    'cstride_size3': hp.choice('cstride_size3',range(1,3)),
    'wb_initializer3': hp.choice('wb_initializer3',initializers),
    'activation3': hp.choice('activation3', activations),
    
    'conv_size4': hp.choice('conv_size4', range(3,6)),
    'out_channels4': hp.choice('out_channels4',range(10,201)),                
    'cstride_size4': hp.choice('cstride_size4',range(1,3)),
    'wb_initializer4': hp.choice('wb_initializer4',initializers),
    'activation4': hp.choice('activation4', activations),

    'conv_size5': hp.choice('conv_size5', range(3,6)),
    'out_channels5': hp.choice('out_channels5',range(10,201)),                
    'cstride_size5': hp.choice('cstride_size5',range(1,3)),
    'wb_initializer5': hp.choice('wb_initializer5',initializers),
    'activation5': hp.choice('activation5', activations),
    'pool_type5': hp.choice('pool_type5',[tf.nn.max_pool,tf.nn.avg_pool]),                    
    'pool_size5': hp.choice('pool_size5',range(2,4)),                      
    'pstride_size5': hp.choice('pstride_size5',range(1,3)), 
    
    'fc_nodes1': hp.choice('fc_nodes1', range(10,2049)),   
    'fc_initializer1': hp.choice('fc_initializer1',initializers),
    'fc_activation1': hp.choice('fc_activation1', activations),
    'keep_prob1': hp.uniform('keep_prob1', 0,1),
    
    'fc_nodes2': hp.choice('fc_nodes2', range(10,2049)),   
    'fc_initializer2': hp.choice('fc_initializer2',initializers),
    'fc_activation2': hp.choice('fc_activation2', activations),
    'keep_prob2': hp.uniform('keep_prob2', 0,1),
    
    'fc_initializer3': hp.choice('fc_initializer3',initializers),

    'optimizer': hp.choice('optimizer',optimizers),
    'batch_size' : hp.choice('batch_size', range(28,256)),
    'learning_rate': hp.uniform('learning_rate', 0.0001,0.1)
        }

def cnn(params):
    print (time.strftime("%Y-%m-%d %H:%M:%S",time.localtime()))
    tf.reset_default_graph()
    image_holder = tf.placeholder(tf.float32, [None,90,90,3])
    label_holder = tf.placeholder(tf.int32, [None]) 
    is_training = tf.placeholder(tf.string) 
    #x_image = tf.reshape(image_holder, [-1,28,28,1])

    #First Conv
    with tf.variable_scope('hidden1'):
        c1 = conv_bn_pool(x = image_holder,conv_size = params['conv_size1'],channels = 3, 
                                    out_channels = params['out_channels1'],cstride_size = params['cstride_size1'],
                                    wb_initializer = params['wb_initializer1'],activation = params['activation1'],
                                    pool_type = params['pool_type1'],pool_size = params['pool_size1'],
                                    pstride_size = params['pstride_size1'],is_training = is_training)
    #Second variable_scope
    with tf.variable_scope('hidden2'):
        c2 = conv_bn_pool(x = c1,conv_size = params['conv_size2'],channels = params['out_channels1'], 
                                    out_channels = params['out_channels2'],cstride_size = params['cstride_size2'],
                                    wb_initializer = params['wb_initializer2'],activation = params['activation2'],
                                    pool_type = params['pool_type2'],pool_size = params['pool_size2'],
                                    pstride_size = params['pstride_size2'],is_training = is_training)
    #Third variable_scope
    with tf.variable_scope('hidden3'):
        c3 = conv_bn_pool(x =c2,conv_size = params['conv_size3'],channels = params['out_channels2'], 
                                    out_channels = params['out_channels3'],cstride_size = params['cstride_size3'],
                                    wb_initializer = params['wb_initializer3'],activation = params['activation3'])
    #4th variable_scope
    with tf.variable_scope('hidden4'):
        c4 = conv_bn_pool(x =c3,conv_size = params['conv_size4'],channels = params['out_channels3'], 
                                    out_channels = params['out_channels4'],cstride_size = params['cstride_size4'],
                                    wb_initializer = params['wb_initializer4'],activation = params['activation4'])
    #5th variable_scope
    with tf.variable_scope('hidden5'):
        c5 = conv_bn_pool(x = c4,conv_size = params['conv_size5'],channels = params['out_channels4'], 
                                    out_channels = params['out_channels5'],cstride_size = params['cstride_size5'],
                                    wb_initializer = params['wb_initializer5'],activation = params['activation5'],
                                    pool_type = params['pool_type5'],pool_size = params['pool_size5'],
                                    pstride_size = params['pstride_size5'],is_training = is_training)
    #Fully Connected Layer1
    with tf.variable_scope('fully_conn1'):
        c5_flat = tf.reshape(c5, [-1, np.prod(c5.shape[1:])])
        W_fc1 = tf.get_variable(name='weights1', shape=[np.prod(c5.shape[1:]),params['fc_nodes1']],
                                initializer=params['fc_initializer1'])
        b_fc1 = tf.get_variable(name='bias1', shape=[params['fc_nodes1']],initializer=params['fc_initializer1'])
        h_fc1 = params['fc_activation1'](batch_normalization(tf.matmul(c5_flat, W_fc1) + b_fc1,is_training)) 
        h_fc1_drop = tf.nn.dropout(h_fc1, params['keep_prob1'])
    #Fully Connected Layer2
    with tf.variable_scope('fully_conn2'):
        W_fc2 = tf.get_variable(name='weights2',shape=[params['fc_nodes1'],params['fc_nodes2']],
                                initializer=params['fc_initializer2'])
        b_fc2 = tf.get_variable(name='bias2', shape=[params['fc_nodes2']],initializer=params['fc_initializer2'])
        h_fc2 = params['fc_activation2'](batch_normalization(tf.matmul(h_fc1_drop, W_fc2) + b_fc2,is_training)) 
        h_fc2_drop = tf.nn.dropout(h_fc2, params['keep_prob1'])
    #Softmax Layer
    with tf.variable_scope('softmax_layers'):
        W_fc3 = tf.get_variable(name='weights3', shape=[params['fc_nodes2'], 2],
                                initializer=params['fc_initializer3'])
        b_fc3 = tf.get_variable(name='bias3', shape=[2],initializer=params['fc_initializer3'])
        logits = tf.matmul(h_fc2_drop, W_fc3) + b_fc3

    cross_entropy = tf.losses.sparse_softmax_cross_entropy(labels=label_holder, logits=logits)
    loss = tf.reduce_mean(cross_entropy)

    global_step=tf.Variable(0)
    lr=tf.train.exponential_decay(params['learning_rate'],global_step,10,0.95,staircase=True)
    if params['optimizer']=='Adam':
        train_op = tf.train.AdamOptimizer(lr).minimize(loss)
    elif params['optimizer']=='GradientDescend':
        train_op = tf.train.GradientDescentOptimizer(lr).minimize(loss)
    elif params['optimizer']=='Adagrad':
        train_op = tf.train.AdagradOptimizer(0.01).minimize(loss)            
    elif params['optimizer']=='Momentum':
        train_op = tf.train.MomentumOptimizer(lr,0.25).minimize(loss)
    elif params['optimizer']=='Ftrl':
        train_op = tf.train.FtrlOptimizer(lr).minimize(loss)            
    else :
        train_op = tf.train.RMSPropOptimizer(lr).minimize(loss)  
    
    top_k_op = tf.nn.in_top_k(logits, label_holder, 1)
    
    print(params)
    epochs = 15
    train_num_examples=y_train.shape[0]
    max_steps = int(math.ceil(train_num_examples / (params['batch_size']+0.0)))
    total_sample_train = max_steps * params['batch_size']
    best = 0
    wait = 0  #counter for patience
    best_rounds =1
    counter=0
    patience=2
    train_acc_all=[]

    min_fraction_of_examples_in_queue = 0.4
    min_queue_examples = int(y_train.shape[0] * min_fraction_of_examples_in_queue)
    input_queue=tf.train.slice_input_producer([X_train,y_train],shuffle=False)
    X_batch,y_batch = tf.train.batch(input_queue,batch_size=params['batch_size'],num_threads=100,capacity=min_queue_examples + 3 * params['batch_size'])  
    # 
    sess = tf.InteractiveSession()
    tf.global_variables_initializer().run()
    # 
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(coord=coord)
    for epoch in xrange(epochs):
        #print ("epoch--:%d"%(epoch+1))        
        true_count_train = 0
        for step in xrange(max_steps):
            image_batch, label_batch = sess.run([X_batch,y_batch])
            feed_dict ={image_holder: image_batch,label_holder: label_batch, is_training: 'True'}
            #Train
            _, l, a = sess.run([train_op, loss,top_k_op],feed_dict=feed_dict)
            '''if step%100 == 0:
                print('Batch_size:',params['batch_size'], 'Train Batch Accuracy:',np.sum(a)/ params['batch_size'],'step:',step+1)'''
            
            true_count_train += np.sum(a)
        train_accuracy = true_count_train / total_sample_train
        train_acc_all.append(train_accuracy)

        accuracy = sess.run([top_k_op],feed_dict={image_holder: X_val,label_holder: y_val, is_training: 'False'})
        true_count = np.sum(accuracy)
        current_accuracy = true_count / y_val.shape[0]
        counter +=1

        if current_accuracy > best:
            best = current_accuracy
            best_rounds=counter
            wait = 0
        else:
            wait += 1 #incremental the number of times without improvement
            if wait >= patience : #no more patience, retrieve best model
                break
    print ("epoch--:%d"%(epoch+1))
    print('Best number of rounds: %d , best_val_precision: %f ,best_train_accuracy: %f' % (best_rounds, best,train_acc_all[best_rounds-1]))
    coord.request_stop()
    coord.join(threads)
    sess.close()        
    return {'loss': current_accuracy*(-1),'val_accuracy':best,'train_accuracy':train_acc_all[best_rounds-1],'best_rounds':best_rounds,'status': STATUS_OK}


print (time.strftime("%Y-%m-%d %H:%M:%S",time.localtime()))
max_evals=100
doe_num=70
trials = Trials()
best =  fmin(cnn, space, algo=tpe.suggest, trials=trials, verbose=0, max_evals=max_evals, doe_num=doe_num)
print ("best:",best)
print (trials.best_trial)
print (time.strftime("%Y-%m-%d %H:%M:%S",time.localtime()))
```
